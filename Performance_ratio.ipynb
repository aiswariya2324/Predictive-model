```python
# Import necessary libraries
import numpy as np

# Define system parameters
expected_energy_production = 1000  # Expected energy production in kWh
actual_energy_production = 700    # Actual energy production in kWh
solar_irradiance = 1200            # Solar irradiance in W/m^2
module_area = 10                   # Module area in square meters
system_capacity = 5                # System capacity in kW

# Calculate the performance ratio
def calculate_performance_ratio(expected_energy, actual_energy, irradiance, area, capacity):
    expected_energy_output = capacity * area * irradiance / 1000  # Convert irradiance to kW/m^2
    performance_ratio = actual_energy / expected_energy_output
    return performance_ratio

# Calculate and print the performance ratio
pr = calculate_performance_ratio(expected_energy_production, actual_energy_production, solar_irradiance, module_area, system_capacity)
print(f"Performance Ratio: {pr:.2f}")
```
